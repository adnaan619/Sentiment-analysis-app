```python
#Now lets preprocess this dataset
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
#Downloading the necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
#Function for cleaning and preprocessing the reviewText and the summary
def preprocess_text(text):
    #If text is NaN, return an empty string
    if pd.isna(text):
        return ""
    
    #Convert text to lowercase
    text = text.lower()
    
    #Remove the punctuation and any special characters 
    text = re.sub(r'[^a-z0-9\s]', '', text)
    
    #Tokenize text
    tokens = word_tokenize(text)
    
    #Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    #Initialize the WordLemmatizer
    lemmatizer = WordNetLemmatizer()
    
    #Lemmatize tokens 
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    #Reconstruct the text
    text = ' '.join(tokens)
    return text

#Apply the preprocessing function to the reviewText and summary columns
df['clean_reviewText'] = df['reviewText'].apply(preprocess_text)
df['clean_summary'] = df['summary'].apply(preprocess_text)
```

```python
#Now lets reduce the dataset to 200,000 reviews while maintaining rating diversity
# Calculate the number of reviews to sample from each rating class based on proportion
ratings_count = df_top_reviews['overall'].value_counts(normalize=True) * 200000
```

```python
# Since the number of reviews to sample should be an integer, we round the count
ratings_count = ratings_count.round().astype(int)
```

```python
# Create a new DataFrame to hold the stratified sample
df_stratified = pd.DataFrame()
```

```python
# Stratified sampling from each rating class
for rating, count in ratings_count.items():
    df_rating = df_top_reviews[df_top_reviews['overall'] == rating].sample(n=count, random_state=1)
    df_stratified = pd.concat([df_stratified, df_rating])
```

```python
df_stratified.info()
```

```python
#This method below is to handle stratified sampling to get exactly 200k reviews
```

```python
# Step 1: Determine proportions
proportions = df_top_reviews['overall'].value_counts(normalize=True)
```

```python
# Step 2: Calculate the number of samples, targeting exactly 200,000
sample_sizes = (proportions * 200000).astype(int)
```

```python
# Adjust if the sum of sample sizes does not equal 200,000 due to rounding
while sample_sizes.sum() != 200000:
    difference = 200000 - sample_sizes.sum()
    if difference > 0:
        sample_sizes[sample_sizes.idxmin()] += difference
    else:
        sample_sizes[sample_sizes.idxmax()] += difference  # difference is negative
```

```python
#Step 3 perform stratified sampling
df_amazon_fashion = df_top_reviews.groupby('overall').apply(
    lambda x: x.sample(n=sample_sizes[x.name])
).reset_index(drop=True)
```
